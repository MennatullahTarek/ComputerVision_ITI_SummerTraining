###  <span style="color:green;"> Q)</span> Generate an `RGB` image `250*250` and make `every single row` with different random color.
- make an `RGB` image `250*250`.
- make `every single row` with different `random`.


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from random import randint
```

```python
RGB_image=np.zeros((250,250,3),np.uint8)
```

```python

for i in range (0,249):
    RGB_image[i: ] = np.random.randint(0, 255, size=3) 

# Display the modified image
plt.imshow(RGB_image[:,:,::-1])
```
