###  <span style="color:green;"> Q)</span> apply on the following image the `shown filter`.
- `Load` the image.
- apply `sharpen` filter.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = cv2.imread('blur.png')
plt.imshow(image[:,:,::-1]) 
```

```python
kernel_sharpen = np.array([[0 , -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

sharp_image = cv2.filter2D(image, -1, kernel_sharpen)

f, axes = plt.subplots(1, 2, figsize=(25,15))
axes[0].imshow(image[:,:,::-1]), axes[0].set_title('Original')
axes[1].imshow(sharp_image[:,:,::-1]), axes[1].set_title('Sharpen filter')
```
