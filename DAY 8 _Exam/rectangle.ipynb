 ###  <span style="color:green;"> Q)</span> `load` the following image then when you start `single click` in left mouse button set `start point` of rectangle and when move mouse cursor to another point and `double click` on left mouse button set `end point` then draw `rectangle` from started point to end point then should write `Rectangle{counter}`below the image based on counter value.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('viola.jpg')
plt.imshow(image[:,:,::-1]) 
```

```python
count=0
def call_point_1(events1,x1,y1,flag1,param1):  
    if events1== cv2.EVENT_LBUTTONDOWN:
         print(x1,y1)
    return (x1,y1)
   

def call_back_point_2(events2,x2,y2,flag2,param2):
    if events2== cv2.EVENT_LBUTTONDBLCLK:
        global count
        print(x2,y2)
        count += 1
    return (x2,y2)
```

```python

cv2.namedWindow('image')

point2 = cv2.setMouseCallback('image',call_back_point_2)


while 1:
    cv2.imshow('image',image)
    k=cv2.waitKey(1)
    if k== ord('q'):
        break
cv2.destroyAllWindows()
```

```python

cv2.namedWindow('image')

point1 = cv2.setMouseCallback('image',call_point_1)



while 1:
    cv2.imshow('image',image)
    k=cv2.waitKey(1)
    if k== ord('q'):
        break
cv2.destroyAllWindows()
```

```python
# Draw red rectangle using the coordinates
cv2.rectangle(image, point1, point2 , (0, 0, 0), 2)

# Write "Cat" below the rectangle with yellow color
cv2.putText(image,f'Rectangle {count}', (100,380),cv2.FONT_HERSHEY_PLAIN,2,(0,255,255),2,cv2.LINE_AA)

plt.imshow(image[:,:,::-1])
```
